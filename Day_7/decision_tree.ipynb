```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import confusion_matrix, accuracy_score

# set style for plots
sns.set_style('whitegrid')
```

```python
# load the dataset
data = pd.read_csv('Social_Network_Ads.csv')
data.head()
```

2. Preprocessing

we will use Age and EstimateSalary as our features(X) and Purchased as our target variable (Y)

```python
X = data.iloc[:,[2,3]].values
Y = data.iloc[:,4].values

# split the dataset into the training and test set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

3. Feature Scaling

although Decision Trees do not strictly require freature scaling, it can be helpful for visulisation and comparison with other models.

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

4. Training the Decision Tree Model

```python
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0,max_depth=3)
classifier.fit(X_train,Y_train)
```

5. Model Evaluation

```python
y_pred = classifier.predict(X_test)

cm = confusion_matrix(Y_test,y_pred)
print("Confusion Matrix:")
print(cm)
print(f"Accuracy Score: {accuracy_score(Y_test,y_pred):.2f}")
```

6. visualizing the Decision Tree

```python
plt.figure(figsize=(20,10))
plot_tree(classifier,
            feature_names=['Age','Estimated Salary'],
            class_names=['Not Purchased','Purchased'],
            filled=True,
            rounded=True,
            fontsize=12)
plt.title("Decision Tree Visualization")
plt.show()
```

7. Visualizing the Decision Boundary

```python
from matplotlib.colors import ListedColormap

X_set, y_set = X_test, Y_test
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))

plt.figure(figsize=(10,6))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('red', 'blue'))(i), label=j)
    
plt.title('Decision Tree Classification (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
