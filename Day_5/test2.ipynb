Simple Logistics regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

sns.set_theme(style="darkgrid")
```

```python
# load the simple dataset
df = pd.read_csv("simple_logistics.csv")
print("Datatset:")
print(df.head())

# Visualize the data points
plt.figure(figsize=(8, 5))
plt.scatter(df['Hours'], df['Pass'], c=df['Pass'], cmap='plasma', edgecolors='k')
plt.xlabel('Hours Studied')
plt.ylabel('Pass (1) / Fail (0)')
plt.title('Hours vs Result')
plt.show()
```

```python
# prepare features and target variable
X = df[['Hours']] #input feature
y = df['Pass'] #output variable

#initialize and train the model
model = LogisticRegression()
model.fit(X, y)

print("Model trained successfully.")
```

```python
#visualizing the sigmoid curve (Logistic regression curve)
plt.figure(figsize=(10, 6))
plt.scatter(df['Hours'], df['Pass'], c= df['Pass'], cmap='bwr', label='Data Points')

#Generate a range of values for the curve
X_test_range = np.linspace(0,10, 100).reshape(-1, 1)
y_pred_range = model.predict_proba(X_test_range)[:, 1]

plt.plot(X_test_range, y_pred_range, color='green',linewidth=3, label='Logistic Curve')
plt.axhline(y=0.5, color='gray', linestyle='--', label='Decision Boundary (0.5)')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression Sigmoid Curve')
plt.legend()
plt.show()
```

```python
try:
    # Get user input
    user_hours = float(input("Enter study hours (0-10): "))

    #prepare input for model
    X_user = np.array([[user_hours]])

    #make prediction
    prediction = model.predict(X_user)[0]
    probability = model.predict_proba(X_user)[0][1]

    #output result
    if prediction == 1:
        print(f"\nResult: PASS ✅")
        print(f"confidence: {probability*100:.2f}%")
    else:
        print(f"\nResult: FAIL ❌")
        print(f"confidence: {(1-probability)*100:.2f}%")
except ValueError:
    print("Invalid input. Please enter a numerical value for hours.")
```
